# Regridding the Data

#### The original data came in a Gaussian Grid format. This needed to be converted to a lat/lon so that it would be plotted and sliced in that dimension. 

```python
# Regridding 
```

```python
def mirror_point_at_360(ds):
  extra_point = (
      ds.where(ds.longitude == 0, drop=True)
      .assign_coords(longitude=lambda x: x.longitude + 360)
  )
  return xr.concat([ds, extra_point], dim='values')

def build_triangulation(x, y):
  grid = np.stack([x, y], axis=1)
  return scipy.spatial.Delaunay(grid)

def interpolate(data, tri, mesh):
  indices = tri.find_simplex(mesh)
  ndim = tri.transform.shape[-1]
  T_inv = tri.transform[indices, :ndim, :]
  r = tri.transform[indices, ndim, :]
  c = np.einsum('...ij,...j', T_inv, mesh - r)
  c = np.concatenate([c, 1 - c.sum(axis=-1, keepdims=True)], axis=-1)
  result = np.einsum('...i,...i', data[:, tri.simplices[indices]], c)
  return np.where(indices == -1, np.nan, result)
```

```python
#Select Date Range 1979 - 2020
rs_1520 = seasonal.pipe(mirror_point_at_360)
tri = build_triangulation(rs_1520.longitude, rs_1520.latitude)
```

```python
longitude = np.linspace(0, 360, num=360*4+1)
latitude = np.linspace(-90, 90, num=180*4+1)
mesh = np.stack(np.meshgrid(longitude, latitude, indexing='ij'), axis=-1)
```

```python
#Regridding Surface Sea Temperature 
sst_mesh = interpolate(rs_1520.sst.values, tri, mesh)
sst_1520 = xr.DataArray(sst_mesh, coords=[('time', rs_1520.time.data), ('longitude', longitude), ('latitude', latitude)])
```

```python
# Regridding Soil Moisture
sm_mesh = interpolate(rs_1520.swvl1.values, tri, mesh)
sm_1520 = xr.DataArray(sm_mesh, coords=[('time', rs_1520.time.data), ('longitude', longitude), ('latitude', latitude)])
```

```python
#Apply Land Mask to SM Data 
mask = xr.open_dataset('lsm_1279l4_0.1x0.1.grb_v4_unpack.nc')
masked = mask.lsm 
mask_interp = mask.interp_like(sm_1520[0], method='nearest')
land_mask = mask_interp.lsm[0]
```

```python
#Meaning across time 
lat = sst_1520.latitude
lon = sst_1520.longitude
sst_time_mean = sst_1520.mean(dim=('time'), keep_attrs = True)
sm_time_mean = (sm_1520.mean(dim=('time'), keep_attrs = True)) 
sst = (sst_time_mean.values.T) - 273.15
sm = (sm_time_mean.values.T) 
lat.shape, lon.shape, sst.shape, sm.shape
```

# 1. How similar or different are patterns for sea surface temperature and soil moisture?

By plotting contour plots, we can see if there are zonal patterns and variations that correspond between the sea surface and soil moisture data. Immediately looking at the data,we find that there is a clear zonal coorelation between the sea surface temperature and the soil moisture. 

```python
# Plot the time mean 
fig, ax = plt.subplots(figsize = [20, 18], nrows= 2)
fig.tight_layout()
#Figure 1 
levels = np.arange(-10, 45, 5)
sst_map = ax[0].contourf(lon, lat, sst, levels, cmap = "coolwarm", extend = "both")
#ax[0].contour(lon, lat, temp, levels=[0], colors = 'white', extend = "both")
ax[0].set_xticks([])
#ax[0].set_xlabel('Longitude')
ax[0].set_ylabel('Latitude')
color_bar_ratio = sst.shape[0]/sst.shape[1] 
ax[0].set_title("Sea Surface Temperature Mean From 2015 to 2020", fontsize = 20)
plt.colorbar(sst_map , ax =ax[0], label ='\N{DEGREE SIGN}C', shrink = 0.7)

levels2 = np.arange(0.0, 0.5, 0.01)
sm_map = ax[1].contourf(lon, lat, sm, levels2, cmap = "Blues", extend = "both")
#ax[1].contour(lon, lat, sm, levels=[0.0], colors = 'white', extend = "both")
ax[1].set_xticks([0, 60, 120, 180, 240, 300])
ax[1].set_xlabel('Longitude')
ax[1].set_ylabel('Latitude')
color_bar_ratio2 = sm.shape[0]/sm.shape[1] 
ax[1].set_title("Soil Moisture Mean From 2015 to 2020", fontsize = 20)
plt.colorbar(sm_map, ax =ax[1], label ='m$^3$', shrink = 0.7)
plt.show() 
```

```python
#Group by seasons
sst_s = sst_1520.groupby('time.month')
sm_s = sm_1520.groupby('time.month')
```

```python
# Calculate Seasonal Time Mean
sst_sm = sst_s.mean(dim = 'time')
sm_sm = sm_s.mean(dim = 'time')
```

```python
#Zonal Mean 
sst_zonal = sst_sm.mean(dim='longitude').transpose()
sm_zonal = sm_sm.mean(dim='longitude').transpose()
```

```python
fig, ax = plt.subplots(figsize=(10, 8), ncols=2)
ax[0].contourf(sst_zonal.month, sst_zonal.latitude, sst_zonal.values, cmap = "coolwarm")
ax[1].contourf(sm_zonal.month, sm_zonal.latitude, sm_zonal.values, cmap = "Blues")
ax[0].set_ylabel('Latitude')
ax[1].set_ylabel('Latitude')
ax[0].set_xlabel('Month')
ax[1].set_xlabel('Month')
ax[0].set_title("Zonal Mean of Sea Surface Temperature")
ax[1].set_title("Zonal Mean of Soil Moisture")
plt.colorbar(sst_map , ax =ax[0], label ='\N{DEGREE SIGN}C', shrink = 0.7)
plt.colorbar(sm_map, ax =ax[1], label ='m$^3$', shrink = 0.7)
plt.show()
```

```python
# Calculate Anomaly for Sea Surface Temperature and Soil Moisture
sst_anom2 = (sst_1520 - sst_1520.mean(dim='time'))
sm_anom2 = (sm_1520 - sm_1520.mean(dim='time'))
```

```python
#Pearson Coeffiecienty Between Sea Surface Temperature 2015 and 2020
sst1 = sst_anom2[0].values.flatten()
sstx = sst1[~np.isnan(sst1)]
sst2 = sst_anom2[-1].values.flatten()
ssty = sst2[~np.isnan(sst2)]
```

```python
# correlation test
corr, _ = pearsonr(sstx, ssty)
print('Pearsons correlation: %.3f' % corr)
```
